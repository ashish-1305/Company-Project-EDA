```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=.125,bottom=.1,right=.9,top=.9,wspace=.5,hspace=.2)
plt.subplot(141)
plt.title("gender",fontsize=20)
stu['gender'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(142)
plt.title("race",fontsize=20)
stu['race/ethnicity'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(143)
plt.title("Lunch",fontsize=20)
stu['lunch'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(144)
plt.title("Parental level of education",fontsize=20)
stu['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```
